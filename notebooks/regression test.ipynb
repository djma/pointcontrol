```python
from __future__ import print_function
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
from statsmodels.sandbox.regression.predstd import wls_prediction_std

np.random.seed(9876789) 
```

```python
nsample = 100
x = np.linspace(0, 10, nsample)
X = np.column_stack((x, x**2))
beta = np.array([1, 0.1, 10])
e = np.random.normal(size=nsample)
```

```python
X = sm.add_constant(X)
y = np.dot(X, beta) + e 
```

```python
model = sm.OLS(y, X)
results = model.fit()
print(results.summary())
 
```

```python
results.resid
```

```python
X1 = X[:,[0,1]]
X2 = X[:,[0,2]]


model = sm.OLS(y, X1)
results = model.fit()
print(results.summary())
print(results.resid)

y_residX1 = results.resid
model2 = sm.OLS(y_residX1, X2)
results2 = model2.fit()
print(results2.summary())
print(results2.resid)
```

```python
X1 = X[:,[0]]
X2 = X[:,[1,2]]


model = sm.OLS(y, X1)
results = model.fit()
print(results.summary())
print(results.resid)

y_residX1 = results.resid
model2 = sm.OLS(y_residX1, X2)
results2 = model2.fit()
print(results2.summary())
print(results2.resid)
```
