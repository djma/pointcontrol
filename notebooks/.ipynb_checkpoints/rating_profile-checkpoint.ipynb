```python
import pandas as pd
import ggplot as gg
```

```python
d = pd.read_csv("../datasets/ratings.csv", sep="|", 
                header=None, 
                names=["fencerid", "weapon", "boutid", "ts_mu", "ts_sigma", "prev_ts_mu", "prev_ts_sigma", "start_date"],
                parse_dates=True,
                index_col="start_date")
d = d[d["weapon"] == "Epee"]
d["start_date"] = d.index
f = d.groupby(["fencerid", "start_date"]).last()
fmu = f["ts_mu"].unstack(0).fillna(method="pad")
fmu365 = f["ts_mu"].unstack(0).fillna(method="pad", limit=365)
d.head(10)
```

```python
fmu.mean(axis=1).plot()
```

```python
fmu365.median(axis=1).plot(title="Median rating of active epeeists").set_xlabel("as of date")
```

```python
fmu.std(axis=1).plot()
```

```python
fmu.count(axis=1).plot()
```

```python
fmu365.mean(axis=1).plot()
```

```python
fmu365.std(axis=1).plot()
```

```python
fmu365.count(axis=1).plot()
```

```python
def plot_fencer(fid):
    fencer = d[d["fencerid"] == fid]
    fencer["start_date"] = fencer.index
    fencer["ts_mu_minus_ts_sigma"] = fencer["ts_mu"] - fencer["ts_sigma"]
    fencer["ts_mu_plus_ts_sigma"] = fencer["ts_mu"] + fencer["ts_sigma"]
    p = gg.ggplot(gg.aes(x="start_date", y="ts_mu"), data=fencer)
    #print fencer
    return p + gg.geom_line() + gg.geom_area(gg.aes(ymin="ts_mu_minus_ts_sigma",ymax="ts_mu_plus_ts_sigma"), data=fencer, fill="grey", alpha="0.5")

```

```python
plot_fencer(35742)
```

```python
plot_fencer(8718)
```

```python
plot_fencer(7728)
```
