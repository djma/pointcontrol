```python
import pandas as pd
import ggplot as gg
```

```python
d = pd.read_csv("../datasets/epee_bouts.txt", sep="|", 
                header=None, 
                names=["boutid", "eventid", "fencer1id", "fencer2id", "score1", "score2", "type", "start_date"],
                parse_dates=True)
d.index = pd.to_datetime(d["start_date"], format='%Y-%m-%d')
d.head()
```

```python
s = d.groupby([pd.TimeGrouper(freq='M'), "type"]).count()
#s["start_date"] = s.index
#s["start_date"] = s["start_date"].map(lambda x : x if x.month % 3 == 0 else "")
s = s.unstack(1)
s["boutid"].head()
p = s["boutid"].plot(figsize=(12,8), title="Number of bouts by month (stacked)", kind="bar", width=1, grid=False, stacked=True)
```

```python
d2 = d.copy(deep=True)
d2["fencer1id"] = d["fencer2id"]
d2["fencer2id"] = d["fencer1id"]
d2["score1"] = d["score2"]
d2["score2"] = d["score2"]
d3 = pd.concat([d,d2])
```

```python
s = d3.groupby([pd.TimeGrouper(freq='A'), "fencer1id"]).count()
s["count"] = s["boutid"]
s = s.unstack(0)
p = s["count"]

p = p.stack("start_date").hist(bins=200, figsize=(6,4))
p.set_xlim((0,150))
p.set_title("Histogram of the number of bouts per fencer per year")
p.set_xlabel("Number of bouts per year")
p.set_ylabel("Fencer-year counts")
```

```python
p = d.groupby("eventid").count()["boutid"].hist(bins=200, figsize=(6,4))
p.set_xlim((0,200))
p.set_title("Histogram of the number of bouts per tournament event")
p.set_xlabel("Number of bouts per tournament event")
p.set_ylabel("Counts")
```
